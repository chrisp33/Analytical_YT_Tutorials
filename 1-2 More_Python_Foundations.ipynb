# Reusing / Reusable Code

## Building Functions

```python
# This code snippet converts a temperature from Celsius to Kelvin.

# The temperature in Celsius to be converted.
c = 20

# Convert the Celsius temperature to Kelvin using the formula: Kelvin = Celsius + 273.15
k = 273.15 + c

# Display the temperature in Kelvin.
k
```

```python
# The previous cell now represented as a function. 

def celsius_to_kelvin(temperature_c):
    """
    Convert temperature from Celsius to Kelvin.
    
    Parameters:
    temperature_c (float or int): Temperature in degrees Celsius.
    
    Returns:
    float: Temperature in Kelvin.
    
    Example:
    >>> celsius_to_kelvin(25)
    298.15
    """
    temperature_k = temperature_c + 273.15  # Conversion formula
    return temperature_k
```

- **Function Definition**: `def celsius_to_kelvin(temperature_c):` declares a function named `celsius_to_kelvin` that takes one argument, `temperature_c`, which represents the temperature in degrees Celsius.
  
- **Docstring**: The triple-quoted string immediately following the function definition provides a detailed description of what the function does, its parameter(s), its return value, and an example of how to use the function. This is vital for anyone who uses your function to understand its purpose and usage without needing to delve into the code details.

- **Conversion Logic**: `temperature_k = temperature_c + 273.15` is where the conversion from Celsius to Kelvin takes place. The constant `273.15` is added to the input temperature in Celsius to convert it to Kelvin.

- **Return Statement**: `return temperature_k` sends the result of the conversion back to the caller. This allows the function's output to be used elsewhere in your code.
---

```python
# Call the function 'celsius_to_kelvin' with an argument of 56 degrees Celsius.
# This will convert the temperature from Celsius to Kelvin and return the result.
# The expected result would be 56 + 273.15 = 329.15 Kelvin.
kelvin_temperature = celsius_to_kelvin(56)

# If you wish to see the result, you can print it:
print(kelvin_temperature)  # Expected output: 329.15
```

```python
# Use the help() function to display the documentation of the 'celsius_to_kelvin' function.
# This will print the docstring of the function, providing details about its purpose,
# parameters, return value, and an example usage. It's a useful way to quickly
# understand how to use a function without having to read its source code.
help(celsius_to_kelvin)
```

## Importing Functions and Libraries

### Math Module

```python
# Import the math module with the alias 'm'. This allows you to access functions,
# constants, and mathematical utilities provided by the module using the prefix 'm.'
import math as m

# Use the log10() function from the math module to calculate the base-10 logarithm
# of 0.001. The base-10 logarithm (log10) of a number is the power to which the number
# 10 must be raised to obtain the value of the input number. For 0.001, which is 10^-3,
# the logarithm base 10 is -3.
log_value = m.log10(0.001)

# If you wish to see the result, you can print it:
print(log_value)  # Expected output: -3.0
```

```python
from math import log10 # we'll cover this in the next video
# Example: Calculate pH values from hydronium ion concentrations
h_concentrations = [0.01, 0.001, 0.0001]  # concentrations in moles per liter
for concentration in h_concentrations:
    pH = -log10(concentration)
    print("pH of solution with H3O+ concentration of", concentration, "M is:", pH)
```

## Matplotlib Module

```python
# Import the matplotlib.pyplot module with the alias 'plt'. This module provides a
# MATLAB-like plotting framework. Using 'plt', you can create a wide range of static,
# animated, and interactive visualizations in Python.
import matplotlib.pyplot as plt

%config InlineBackend.figure_format='retina'  # Improves resolution on large screens
```

```python
# This code snippet is an example of how to plot an absorbance spectrum using matplotlib.
# Absorbance spectra are a common way to represent how much light is absorbed by a sample
# at different wavelengths.

# First, we define two lists: 'wavelengths' and 'absorbance'. These lists represent the
# x-axis (wavelengths in nanometers) and y-axis (absorbance values), respectively, of our plot.
wavelengths = [400, 450, 500, 550, 600, 650, 700]  # in nanometers
absorbance = [0.1, 0.2, 0.4, 0.35, 0.25, 0.15, 0.05]

# Using plt.plot(), we create a line plot with 'wavelengths' on the x-axis and 'absorbance'
# on the y-axis. This line plot visually represents the absorbance spectrum.
plt.plot(wavelengths, absorbance)

# Next, we label the x-axis as 'Wavelength (nm)' using plt.xlabel(). 'nm' stands for nanometers,
# which is a unit of measurement for wavelength.
plt.xlabel('Wavelength (nm)')

# Similarly, we label the y-axis as 'Absorbance' using plt.ylabel(). Absorbance is a unitless
# measure of how much light is absorbed by the sample at each wavelength.
plt.ylabel('Absorbance')

# We give the plot a title 'Absorbance Spectrum' using plt.title() to describe what the plot
# represents.
plt.title('Absorbance Spectrum')

# Finally, plt.show() is called to display the plot. This function renders the plot and shows it
# to us. Without this call, the plot would not be visible in some environments, such as Jupyter
# notebooks or Python scripts run from the command line.
plt.show()
```

## Pandas Module

```python
# Import the Pandas library and alias it as 'pd'. Pandas is a powerful and widely used data analysis 
# and manipulation library in Python. By importing it as 'pd', we can access all its functions, 
# methods, and classes using the 'pd' prefix, which simplifies the syntax for calling Pandas operations.
import pandas as pd
```

```python
# Define lists for wavelengths (in nanometers) and their corresponding absorbance values.
# These might already be defined in your code from previous operations.
wavelengths = [400, 450, 500, 550, 600, 650, 700]  # in nanometers
absorbance = [0.1, 0.2, 0.4, 0.35, 0.25, 0.15, 0.05]

# Create a Pandas Series object named 'spectrum'. A Series is a one-dimensional array-like object
# capable of holding any data type. Here, we use 'wavelengths' as the index, which means each
# absorbance value will be associated with its corresponding wavelength. This is particularly
# useful for data analysis tasks where a clear relationship between two variables is essential.
spectrum = pd.Series(index=wavelengths, data=absorbance)

spectrum
```

```python
# Plot the 'spectrum' Series. Pandas integrates with Matplotlib for easy plotting, using the Series index as the x-axis and values as the y-axis.
spectrum.plot()

# Label the x-axis as 'Wavelength (nm)'.
plt.xlabel('Wavelength (nm)')

# Label the y-axis as 'Absorbance'.
plt.ylabel('Absorbance')

# Set the title of the plot to 'Absorbance Spectrum'.
plt.title('Absorbance Spectrum')

# Display the plot.
plt.show()
```
